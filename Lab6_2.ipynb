**Prompt:** Write a Python class named Student with a constructor to initialize name, roll number, and marks. Include a method display_details() to print student info, take details of student with dynamic input.

```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print(f"Marks: {self.marks}")

# Get dynamic input from the user
name = input("Enter student name: ")
roll_number = input("Enter roll number: ")
marks = input("Enter marks: ")

# Create a Student object with dynamic input
student1 = Student(name, roll_number, marks)

# Display the student details
student1.display_details()
```

**Explaination**:

**class Student**: This line defines a class named Student.

**def __init__(self, name, roll_number, marks):** : This is the constructor. It initializes a Student object with the provided name, roll_number, and marks. self refers to the instance of the class.

**def display_details(self):**: This method prints the name, roll_number, and marks of the Student object.

**name = input("Enter student name: "):** This line prompts the user to enter the student's name and stores the input in the name variable.

**roll_number = input("Enter roll number: "):** This line prompts the user to enter the roll number and stores the input in the roll_number variable.

**marks = input("Enter marks: "):** This line prompts the user to enter the marks and stores the input in the marks variable.

**student1 = Student(name, roll_number, marks):** This line creates a new instance of the Student class named student1, passing the dynamic input values for name, roll number, and marks to the constructor.

**student1.display_details():** This line calls the display_details() method on the student1 object, which prints the student's information.

**Prompt :** Write a Python function to print the first 10 multiples of a number using a for loop, number should be given by user.

```python
def print_multiples_for(number):
  """Prints the first 10 multiples of a number using a for loop."""
  print(f"Multiples of {number}:")
  for i in range(1, 11):
    print(number * i)

# Get input from the user
try:
  num = int(input("Enter a number: "))
  # Example usage:
  print_multiples_for(num)
except ValueError:
  print("Invalid input. Please enter an integer.")
```

**print(number * i):** Inside the loop, this line calculates the current multiple (number multiplied by i) and prints it.

**try:**: This block starts a try...except block to handle potential errors.

**Task Description 3** (Conditional Statements)

• Analyze the generated code

• Ask AI to generate code using other conditional statements

**Prompt :** Write a Python function using nested if-elif-else to classify a person based on age, take age dynamically

```python
def classify_age_nested(age):
  """Classifies a person based on age using nested if-elif-else."""
  if age < 0:
    print("Invalid age.")
  else:
    if age < 13:
      print("Child")
    elif age < 20:
      print("Teenager")
    elif age < 60:
      print("Adult")
    else:
      print("Senior")

# Get dynamic input from the user
try:
  age = int(input("Enter age: "))
  classify_age_nested(age)
except ValueError:
  print("Invalid input. Please enter an integer for age.")
```

**Explaination :**

**Function Definition:** The code defines a function classify_age_nested that takes an age as input.

**Input Validation:** It first checks if the age is less than 0 and prints "Invalid age" if it is.

**Nested Conditional Start:** If the age is not negative, it proceeds to a nested set of conditions within an else block.

**Age Classification (Child/Teenager):** Inside the nested block, it checks if the age is less than 13 ("Child") or less than 20 ("Teenager").

**Age Classification (Adult/Senior):** If the age is not in the child or teenager range, it checks if it's less than 60 ("Adult") or otherwise classifies it as "Senior".

**Dynamic Input and Error Handling:** The code also includes a try-except block to get integer age input from the user and handle cases where the input is not a valid integer.

**Task Description 4** (For and While loops)

• Generate a sum_to_n() function to calculate sum of first n numbers

• Analyze the generated code

• Get suggestions from AI with other controlled looping

**Prompt :** Write a Python function sum_to_n(n) that calculates the sum of first n natural numbers using a for loop, take n value dynamicaly

```python
def sum_to_n(n):
  """Calculates the sum of the first n natural numbers using a for loop."""
  total_sum = 0
  for i in range(1, n + 1):
    total_sum += i
  return total_sum

# Get dynamic input from the user
try:
  num = int(input("Enter a positive integer for n: "))
  if num > 0:
    result = sum_to_n(num)
    print(f"The sum of the first {num} natural numbers is: {result}")
  else:
    print("Please enter a positive integer.")
except ValueError:
  print("Invalid input. Please enter an integer.")
```

**Explaination :**

**Function Definition:** The code defines a function sum_to_n that takes an integer n as input.

**Initialization:** Inside the function, total_sum is initialized to 0 to store the sum.

**For Loop:** A for loop iterates through natural numbers from 1 up to n (inclusive).

**Sum Calculation:** In each loop iteration, the current number is added to total_sum.

**Return Value:** The function returns the final calculated total_sum.


**Dynamic Input and Validation:** The code outside the function takes dynamic integer input for n from the user, validates that it's positive, calls sum_to_n, and handles non-integer input with a try-except block.


**Task Description 5** (Class)

• Use AI to build a BankAccount class with deposit, withdraw, and balance methods.

• Analyze the generated code

• Add comments and explain code

**Prompt :** Create a Python class BankAccount with deposit, withdraw, and check_balance methods, details should be given by user

```python
class BankAccount:
    def __init__(self):
        self.balance = 0
        print("Bank account created with a balance of 0.")

    def deposit(self):
        try:
            amount = float(input("Enter amount to deposit: "))
            if amount > 0:
                self.balance += amount
                print(f"Deposit successful. Current balance: {self.balance}")
            else:
                print("Deposit amount must be positive.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    def withdraw(self):
        try:
            amount = float(input("Enter amount to withdraw: "))
            if amount > 0:
                if self.balance >= amount:
                    self.balance -= amount
                    print(f"Withdrawal successful. Current balance: {self.balance}")
                else:
                    print("Insufficient funds.")
            else:
                print("Withdrawal amount must be positive.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    def check_balance(self):
        print(f"Current balance: {self.balance}")

my_account = BankAccount()
my_account.deposit()
my_account.withdraw()
my_account.check_balance()
```

**Explaination :**

**class BankAccount:** : This line defines a class named BankAccount.

**def __init__(self):**: This is the constructor of the class.

**self.balance = 0:** Initializes the balance attribute of the bank account to 0 when a new BankAccount object is created.

**print("Bank account created with a balance of 0."):** Prints a confirmation message when the account is created.

**def deposit(self):**: This method handles depositing money into the account.
It uses a try-except block to handle potential ValueError if the user enters non-numeric input.

**amount = float(input("Enter amount to deposit: ")):** Prompts the user to enter the deposit amount and converts it to a float.

**if amount > 0:** : Checks if the deposit amount is positive.

**self.balance += amount:** If positive, the amount is added to the balance.

**print(f"Deposit successful. Current balance: {self.balance}"):** Prints a success message and the new balance.

**else: print("Deposit amount must be positive."):** If the amount is not positive, it prints an error message.
**def withdraw(self):** : This method handles withdrawing money from the account.
It also uses a try-except block for input validation.

**amount = float(input("Enter amount to withdraw: ")):** Prompts the user for the withdrawal amount and converts it to a float.

**if amount > 0:** : Checks if the withdrawal amount is positive.
if self.balance >= amount:: If the amount is positive, it checks if there are sufficient funds.

**self.balance -= amount:** If funds are sufficient, the amount is subtracted from the balance.
