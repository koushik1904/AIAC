```python
%%writefile Navigation.js
import React from 'react';
import { View, Text, StyleSheet } from 'react-native';
import { NavigationContainer } from '@react-navigation/native';
import { createNativeStackNavigator } from '@react-navigation/native-stack';

// 1. Define the screen components
const HomeScreen = () => {
  return (
    <View style={styles.screenContainer}>
      <Text style={styles.screenTitle}>Home Screen</Text>
    </View>
  );
};

const AboutScreen = () => {
  return (
    <View style={styles.screenContainer}>
      <Text style={styles.screenTitle}>About Screen</Text>
    </View>
  );
};

const ContactScreen = () => {
  return (
    <View style={styles.screenContainer}>
      <Text style={styles.screenTitle}>Contact Screen</Text>
    </View>
  );
};

// 2. Initialize the Stack Navigator
const Stack = createNativeStackNavigator();

// 3. Define the main Navigation component
const AppNavigator = () => {
  return (
    <NavigationContainer>
      <Stack.Navigator initialRouteName="Home">
        <Stack.Screen name="Home" component={HomeScreen} options={{ title: 'Overview' }} />
        <Stack.Screen name="About" component={AboutScreen} />
        <Stack.Screen name="Contact" component={ContactScreen} />
      </Stack.Navigator>
    </NavigationContainer>
  );
};

const styles = StyleSheet.create({
  screenContainer: {
    flex: 1,
    justifyContent: 'center',
    alignItems: 'center',
    backgroundColor: '#f0f0f0',
  },
  screenTitle: {
    fontSize: 28,
    fontWeight: 'bold',
    color: '#333',
  },
});

export default AppNavigator;
```

```python
%%writefile App.js
import React, { useState, useEffect } from 'react';
import { StyleSheet, Text, View, TextInput, Button, Alert, Dimensions } from 'react-native';
import AsyncStorage from '@react-native-async-storage/async-storage';

const App = () => {
  const [userName, setUserName] = useState('');
  const [greetingName, setGreetingName] = useState('');

  useEffect(() => {
    // Load the name from AsyncStorage when the component mounts
    loadName();
  }, []);

  const loadName = async () => {
    try {
      const storedName = await AsyncStorage.getItem('user_name');
      if (storedName !== null) {
        setGreetingName(storedName);
      }
    } catch (e) {
      console.error('Failed to load name from AsyncStorage', e);
    }
  };

  const saveName = async () => {
    if (userName.trim() === '') {
      Alert.alert('Input Error', 'Please enter your name.');
      return;
    }
    try {
      await AsyncStorage.setItem('user_name', userName);
      setGreetingName(userName);
      setUserName(''); // Clear input after saving
      Alert.alert('Success', 'Name saved successfully!');
    } catch (e) {
      console.error('Failed to save name to AsyncStorage', e);
      Alert.alert('Error', 'Failed to save name.');
    }
  };

  return (
    <View style={styles.container}>
      <Text style={styles.title}>Hello App</Text>

      <TextInput
        style={styles.input}
        placeholder="Enter your name"
        value={userName}
        onChangeText={setUserName}
      />

      <Button title="Save Name" onPress={saveName} />

      {greetingName ? (
        <Text style={styles.greeting}>Hello, {greetingName}!</Text>
      ) : (
        <Text style={styles.greeting}>Hello there!</Text>
      )}
    </View>
  );
};

// Get screen width for responsive styling
const { width } = Dimensions.get('window');

const styles = StyleSheet.create({
  container: {
    flex: 1,
    backgroundColor: '#fff',
    alignItems: 'center',
    justifyContent: 'center',
    padding: 20,
  },
  title: {
    fontSize: 32,
    fontWeight: 'bold',
    marginBottom: 40,
  },
  input: {
    height: 50,
    borderColor: 'gray',
    borderWidth: 1,
    width: width * 0.8, // Make width responsive
    paddingHorizontal: 10,
    marginBottom: 15, // Adjusted for better spacing
    borderRadius: 8,
  },
  greeting: {
    fontSize: 24,
    marginTop: 25, // Adjusted for better spacing
    textAlign: 'center',
  },
});

export default App;
```

EXPLANATION : Certainly! Let's break down the React Native App.js component code piece by piece. This code creates a simple 'Hello App' that takes a user's name, saves it, and displays a personalized greeting, even after the app is closed and reopened, thanks to AsyncStorage. This updated version also includes basic responsiveness for the input field.

App Functional Component:

State Variables:

greetingName: Stores the name that will be displayed in the greeting message.

This hook runs once when the component mounts (due to the empty dependency array []). Its purpose is to call loadName() to retrieve any previously saved name from AsyncStorage.
loadName Async Function:
Asynchronously fetches the value associated with the key 'user_name' from AsyncStorage.
If a name is found, it updates the greetingName state, which in turn updates the displayed greeting.
Includes basic error handling for AsyncStorage operations.
saveName Async Function:
First, it checks if userName is empty or just whitespace. If so, it shows an Alert and stops.
If valid, it asynchronously saves the userName to AsyncStorage under the key 'user_name'.
It then updates greetingName with the newly saved name and clears the userName input field.
It provides feedback to the user with Alert messages for both success and failure.

A conditional Text component displays the greeting: if greetingName has a value, it shows "Hello, [greetingName]!"; otherwise, it defaults to "Hello there!".

This setup ensures that the app looks decent on various screen sizes and maintains the user's name across app sessions.
