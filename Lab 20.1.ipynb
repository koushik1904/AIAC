PROMPT : generate Python login script for input validation vulnerabilities. Instructions: • generate a simple username-password login program. • Review whether input sanitization and validation are implemented. • Suggest secure improvements (e.g., using re for input validation)

```python
def insecure_login():
    print("--- Insecure Login Program ---")
    username = input("Enter username: ")
    password = input("Enter password: ")

    # Simulate a stored username/password for demonstration
    valid_username = "user123"
    valid_password = "pass456"

    if username == valid_username and password == valid_password:
        print("Login successful!")
    else:
        print("Invalid credentials.")

    print("\n--- End Insecure Login Program ---\n")

# Run the insecure login program
insecure_login()
```

EXPLANATION : This code defines a Python function called insecure_login() which demonstrates a very basic and insecure login process. Here's a breakdown:

Prompts for Input: It first prints a header and then uses the input() function to ask the user to enter a username and a password from the console.
Hardcoded Credentials: It has valid_username set to "user123" and valid_password set to "pass456" hardcoded directly within the function. These are the credentials it will check against.
Simple Comparison: It compares the user-provided username and password directly with the hardcoded valid_username and valid_password using a simple == equality check.
Login Result: If both the username and password match exactly, it prints "Login successful!". Otherwise, it prints "Invalid credentials."
Execution: Finally, insecure_login() is called to run this process. This function is intended to show a system without proper input validation, illustrating a vulnerable setup.


Task 2 – SQL Injection Prevention
Task:
Test an AI-generated script that performs SQL queries on a database.
Instructions:
•	Ask AI to generate a Python script using SQLite/MySQL to fetch user details.
•	Identify if the code is vulnerable to SQL injection (e.g., using string concatenation in queries).
•	Refactor using parameterized queries (prepared statements).
Expected Output:
•	A secure database query script resistant to SQL injection


PROMPT : SQL Injection Prevention generate script that performs SQL queries on a database. Instructions: •  generate a Python script using SQLite/MySQL to fetch user details. • Identify if the code is vulnerable to SQL injection (e.g., using string concatenation in queries). • Refactor using parameterized queries (prepared statements). Expected Output: • A secure database query script resistant to SQL injectio

```python
import sqlite3

def create_and_populate_db(db_name=':memory:'):
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()

    # Create users table
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS users (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            username TEXT NOT NULL UNIQUE,
            password TEXT NOT NULL
        )
    ''')

    # Insert sample user records
    try:
        cursor.execute("INSERT INTO users (username, password) VALUES (?, ?)", ('admin', 'adminpass'))
        cursor.execute("INSERT INTO users (username, password) VALUES (?, ?)", ('user1', 'pass123'))
        cursor.execute("INSERT INTO users (username, password) VALUES (?, ?)", ('testuser', 'testpass'))
        conn.commit()
        print("Database 'users' table created and populated successfully.")
    except sqlite3.IntegrityError:
        print("Users already exist in the database.")

    return conn, cursor

def get_user_details_insecure(cursor, username):
    print(f"\nAttempting to fetch details for username: '{username}'")
    # !!! VULNERABLE TO SQL INJECTION !!!
    # The username is directly concatenated into the SQL query string.
    query = f"SELECT id, username FROM users WHERE username = '{username}'"
    print(f"Executing query: {query}")
    try:
        cursor.execute(query)
        results = cursor.fetchall()
        if results:
            print("Query Results:")
            for row in results:
                print(row)
        else:
            print("No users found.")
    except sqlite3.Error as e:
        print(f"An SQL error occurred: {e}")

# Main execution
conn, cursor = create_and_populate_db()

print("\n--- Demonstrating Insecure Query ---")

# 1. Call with a legitimate username
get_user_details_insecure(cursor, 'admin')

# 2. Call with a malicious SQL injection string
# This payload will bypass authentication or expose all users.
# The ' OR '1'='1' part makes the WHERE clause always true.
get_user_details_insecure(cursor, "' OR '1'='1")

# 3. Another malicious example for different behavior
# This might list all users without a specific password check.
get_user_details_insecure(cursor, "' OR 1=1 --")

# Close the database connection
conn.close()
print("\nDatabase connection closed.")
```

EXPLANATION : This Python script, using the sqlite3 library, demonstrates how a simple database query can be vulnerable to SQL injection. It first sets up an in-memory SQLite database with a users table and some sample entries. The core of the demonstration is the get_user_details_insecure function. This function constructs its SQL query by directly concatenating the user-provided username into the SQL string. This is the critical vulnerability point.

The script then proceeds to call this insecure function with three different inputs:

A legitimate username, which works as expected.
A malicious string ' OR '1'='1', which causes the WHERE clause to always evaluate to true, effectively bypassing any intended username check and returning all users.
Another malicious string ' OR 1=1 --', which similarly returns all users by making the WHERE clause true and using -- to comment out the rest of the original query.
The output clearly shows how the malicious inputs successfully extract data that should not be accessible, highlighting the SQL injection vulnerability.

Task:

PROMPT :Cross-Site Scripting (XSS) -generate HTML form with JavaScript for XSS vulnerabilities. Instructions: •  generate a feedback form with JavaScript-based output. • Test whether untrusted inputs are directly rendered without escaping. • Implement secure measures (e.g., escaping HTML entities, using CSP). Expected Output: • A secure form that prevents XSS attacks



```python
html_content = '''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Insecure Feedback Form (XSS Demo)</title>
    <style>
        body { font-family: Arial, sans-serif; margin: 20px; }
        .container { max-width: 600px; margin: auto; padding: 20px; border: 1px solid #ccc; border-radius: 8px; }
        label, textarea, button { display: block; margin-bottom: 10px; }
        textarea { width: 100%; height: 100px; padding: 8px; border: 1px solid #ddd; border-radius: 4px; }
        button { padding: 10px 15px; background-color: #007bff; color: white; border: none; border-radius: 4px; cursor: pointer; }
        button:hover { background-color: #0056b3; }
        #feedbackOutput { margin-top: 20px; padding: 10px; border: 1px dashed #f00; background-color: #ffe0e0; }
    </style>
</head>
<body>
    <div class="container">
        <h1>Feedback Form</h1>
        <p>This form is intentionally vulnerable to XSS for demonstration purposes.</p>

        <form id="feedbackForm">
            <label for="comment">Your Feedback:</label>
            <textarea id="comment" name="comment" placeholder="Enter your feedback here..."></textarea>
            <button type="submit">Submit Feedback</button>
        </form>

        <h2>Submitted Feedback:</h2>
        <div id="feedbackOutput">No feedback submitted yet.</div>
    </div>

    <script>
        document.getElementById('feedbackForm').addEventListener('submit', function(event) {
            event.preventDefault(); // Prevent default form submission

            const commentInput = document.getElementById('comment').value;
            const feedbackOutputDiv = document.getElementById('feedbackOutput');

            // !!! VULNERABLE TO XSS !!!
            // Directly injecting user input using innerHTML without sanitization
            feedbackOutputDiv.innerHTML = 'You said: ' + commentInput;

            document.getElementById('comment').value = ''; // Clear the input field
        });
    </script>
</body>
</html>
'''

# Save the HTML content to a file
with open('insecure_xss_form.html', 'w') as f:
    f.write(html_content)

print("Generated 'insecure_xss_form.html' which contains an XSS-vulnerable feedback form.")
```

EXPLANATION : This code generates an HTML file named insecure_xss_form.html. This file creates a simple feedback form with some basic styling. The form includes a text area for user comments and a submit button. The crucial part for the demonstration of XSS is the embedded JavaScript. When the form is submitted, the script takes the user's input directly from the text area and injects it into a div element on the page using innerHTML without any sanitization. This direct injection of untrusted user input makes the form vulnerable to Cross-Site Scripting (XSS) attacks, as a malicious script entered by the user would be executed by the browser.

Scenario:
Students pick an AI-generated project snippet (e.g., login form, API integration, or file upload).
Instructions:

•	Prompt AI to suggest secure coding practices to fix issues.
•	Compare insecure vs secure versions side by side.
Expected Output:
•	A security-audited code snippet with documented vulnerabilities and fixes.


PROMPT :  Real-Time Application: Security Audit of AI-Generated Code Scenario: Students pick an AI-generated project snippet (e.g., login form, API integration, or file upload). Instructions: • Perform a security audit to detect possible vulnerabilities. • Prompt AI to suggest secure coding practices to fix issues. • Compare insecure vs secure versions side by side. Expected Output: • A security-audited code snippet with documented vulnerabilities and fixes.

```python
html_content_insecure = html_content # Using the previously defined insecure content

# Replace innerHTML with textContent for security
secure_html_content = html_content_insecure.replace(
    "feedbackOutputDiv.innerHTML = 'You said: ' + commentInput;",
    "feedbackOutputDiv.textContent = 'You said: ' + commentInput;"
)

# Save the secure HTML content to a new file
with open('secure_xss_form.html', 'w') as f:
    f.write(secure_html_content)

print("Generated 'secure_xss_form.html' with XSS protection using textContent.")
```

EXPLANATION : This code snippet generates a secure version of the HTML content that was previously created. It addresses the Cross-Site Scripting (XSS) vulnerability by replacing the use of innerHTML with textContent. While innerHTML can interpret and execute HTML or JavaScript provided by a user, textContent treats all input as plain text. This prevents any malicious scripts entered by a user from being executed in the browser, making the form secure against XSS attacks.
